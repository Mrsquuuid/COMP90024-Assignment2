# COVID-19 AU Monthly Cases



```python
import pandas as pd

df = pd.read_csv("COVID/cases_daily_aus.csv")
df.head()
```

```python
df.info()
```

```python
df[["Day", "Month", "Year"]] = pd.DataFrame(df["Date"].str.split("/").tolist()).astype(int)
for col in ["Date", "Day"]:
    del df[col]
df.head()
```

```python
gb = df.groupby(["Month", "Year"])
out = gb.sum().sort_values(["Year", "Month"])
out
```

```python
out.to_csv("COVID/out.csv")
df = pd.read_csv("COVID/out.csv")
df
```
